```python
#check if there are mixed data types 
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Change the datatype of the num_students column into float. For a Pandas dataframe with a column of numbers with commas, use this:
df["num_students"] = [float(str(i).replace(",", "")) for i in df["num_students"]]
```

```python
# Change the datatype of the international_students column into float. For a Pandas dataframe with a column of numbers with percentage, use this:
df["international_students"] = [float(str(i).replace("%", "")) for i in df["international_students"]]
```

```python
#change the datatype of the female_male_ratio column 
df['female_male_ratio'] = df['female_male_ratio'].astype(str)
```

```python
#check if there are still mixed data types 
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```
